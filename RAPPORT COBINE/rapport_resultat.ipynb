```python
import matplotlib.pyplot as plt

notes = [12, 14, 15, 13, 16, 17, 11, 10, 12, 14, 15, 13, 15, 16, 11, 12, 13, 14, 17, 18]

# Créer une figure avec deux sous-graphiques
fig, axs = plt.subplots(1, 2, figsize=(8, 6))

# Histogramme
axs[0].hist(notes, bins=8, edgecolor='black', color='skyblue')
axs[0].set_xlabel('Notes')
axs[0].set_ylabel('Nombre d\'élèves')
axs[0].set_title('Histogramme des notes')
axs[0].grid(True)

# Préparer les données pour le graphique à secteurs
unique_notes = list(set(notes))
note_counts = [notes.count(note) for note in unique_notes]

# Graphique à secteurs
axs[1].pie(note_counts, labels=unique_notes, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
axs[1].set_title('Graphique à secteurs des notes')

# Afficher la figure
plt.tight_layout()
plt.show()
```
